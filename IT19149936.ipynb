```python
#Drop Columns
```

```python
data = data.drop(columns=['Income','Education','PhysActivity','NoDocbcCost','GenHlth','MentHlth','AnyHealthcare','Fruits','Veggies','HeartDiseaseorAttack','HvyAlcoholConsump','CholCheck','Stroke','PhysHlth'])
```

```python
#Split the Dataset
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.20)
```

```python
X_train.shape, X_test.shape
```

```python
data.info()
```

```python
#Training andTesting Accuracy
```

```python
def model_acc(model):
    model.fit(X_train,Y_train)
    acc1 = model.score(X_train,Y_train)
    acc2 = model.score(X_test,Y_test)
    print(str(model)+' Training Accuracy --> '+str(acc1))
    print(str(model)+' Testing Accuracy --> '+str(acc2))
```

```python
from sklearn.naive_bayes import GaussianNB
GNBmodel = GaussianNB()
model_acc(GNBmodel)
```

```python
from sklearn.naive_bayes import MultinomialNB
MNBmodel = MultinomialNB()
model_acc(MNBmodel)
```

```python
from sklearn.linear_model import SGDClassifier
SGDCmodel = SGDClassifier()
model_acc(SGDCmodel)
```

```python
from xgboost.sklearn import XGBClassifier
XGBmodel = XGBClassifier()
model_acc(XGBmodel)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
GBCmodel = GradientBoostingClassifier()
model_acc(GBCmodel)
```

```python
from sklearn.tree import DecisionTreeClassifier
DTmodel = DecisionTreeClassifier()
model_acc(DTmodel)
```

```python
from sklearn.ensemble import RandomForestClassifier
RFmodel = RandomForestClassifier()
model_acc(RFmodel)
```

```python
#Cross Validation for Selected Algorithm
```

```python
from sklearn.model_selection import cross_val_score
acc1 = cross_val_score(RFmodel,X_train,Y_train, cv=5).mean()
acc2 = cross_val_score(RFmodel,X_test,Y_test, cv=5).mean()
print(str(RFmodel)+' Training Accuracy --> '+str(acc1))
print(str(RFmodel)+' Testing Accuracy --> '+str(acc2))
```

```python
#Hyperparameter tuning
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'criterion': ['gini','entropy','log_loss'],
    'max_depth' : [1,2,3,0]
}
grid_search = GridSearchCV(estimator=RFmodel,param_grid=param_grid)
```
